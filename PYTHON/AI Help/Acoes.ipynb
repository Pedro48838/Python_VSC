```python
import pandas as pd
import pandas_datareader as pdr
import matplotlib.pyplot as plt
import numpy as np
import os
import yfinance as yf
```

```python

# Escolha uma ação para análise
acao = "BBDC4.SA"

# Defina o intervalo de datas para análise
start_date = "2019-01-01"
end_date = "2021-12-31"

# Baixe os dados históricos da ação selecionada
df = yf.download(acao, start=start_date, end=end_date)

# Imprima o DataFrame resultante
print(df)

```

```python
# Visualize os dados
df['Close'].plot()
plt.title('Preço de fechamento das ações')
plt.xlabel('Data')
plt.ylabel('Preço')
plt.show()
```

```python
# Calcule o retorno diário da ação
df['Daily Return'] = df['Close'].pct_change()
```

```python
# Calcule a média móvel de 50 dias do preço de fechamento das ações
df['50 Day MA'] = df['Close'].rolling(window=50).mean()
```

```python
# Calcule a média móvel de 200 dias do preço de fechamento das ações
df['200 Day MA'] = df['Close'].rolling(window=200).mean()
```

```python
# Visualize as médias móveis
df[['Close', '50 Day MA', '200 Day MA']].plot()
plt.title('Médias móveis do preço de fechamento das ações')
plt.xlabel('Data')
plt.ylabel('Preço')
plt.show()
```

```python
# Calcule a volatilidade anualizada da ação
retorno_anual = df['Daily Return'].mean() * 252
volatilidade_anual = df['Daily Return'].std() * np.sqrt(252)
print(f'Retorno anual: {retorno_anual}')
print(f'Volatilidade anual: {volatilidade_anual}')
```
